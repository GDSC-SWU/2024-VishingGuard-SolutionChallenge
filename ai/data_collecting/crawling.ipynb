### 보이스피싱 텍스트 크롤링

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
from urllib.parse import urljoin  # urljoin 함수 추가

def get_대출사기형_info():
    list_url = []     # 주소 리스트
    list_content = []     # 본문 리스트

    sleeptime = 0.5     # 프로세스 일시 정지 시간

    # 페이지를 돌며 게시글 주소 크롤링
    for page in range(1, 20, 1):
        try:
            # 페이지 이동
            url = f'https://www.fss.or.kr/fss/bbs/B0000206/list.do?menuNo=200690&page={page}'
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')
            time.sleep(sleeptime)

            # 주소 크롤링
            current = soup.select('#content > div.bd-list > table > tbody > tr > td.title > a')
            for url_tag in current:
                # 절대 경로로 변환하여 리스트에 추가
                absolute_url = urljoin(url, url_tag['href'])
                list_url.append(absolute_url)
        except Exception as e:
            print(f"An error occurred while crawling URLs: {e}")

    # 게시글을 돌며 텍스트 크롤링
    for url in list_url:
        try:
            # 게시물 이동
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')
            time.sleep(sleeptime)

            # 텍스트 크롤링
            cur_content = []
            current = soup.select('#content > div.bd-view > div > div > p')


            # 게시글에 텍스트가 없는 경우
            if not current:
                list_content.append(cur_content)
                continue

            # 게시글에 텍스트가 있는 경우
            for sentence in current:
                cur_content.append(sentence.get_text(strip=True))
            list_content.append(cur_content)
        except Exception as e:
            print(f"An error occurred while crawling content: {e}")

    voicephishing_대출사기형 = pd.DataFrame({'주소': list_url, '내용': list_content})
    return voicephishing_대출사기형

# 함수 호출 및 결과 출력
result_df = get_대출사기형_info()
print(result_df)
```

```python
# 텍스트 있는 경우만 남김
voicephishing_대출사기형_text = list(voicephishing_대출사기형['내용'])
voicephishing_대출사기형_text = [i for i in voicephishing_대출사기형_text if i != '[]']
len_대출사기형_text = len(voicephishing_대출사기형_text)
print(f' 대출사기형 보이스피싱 텍스트 건수: {len_대출사기형_text}')

voicephishing_수사기관사칭형_text = list(voicephishing_수사기관사칭형['내용'])
voicephishing_수사기관사칭형_text = [i for i in voicephishing_수사기관사칭형_text if i != '[]']
len_수사기관사칭형_text = len(voicephishing_수사기관사칭형_text)
print(f' 수사기관사칭형 보이스피싱 텍스트 건수: {len_수사기관사칭형_text}')


# 문자열의 리스트화
for i in range(len_대출사기형_text):
    cur_text = voicephishing_대출사기형_text[i]
    cur_text = ast.literal_eval(cur_text)
    voicephishing_대출사기형_text[i] = cur_text
    
for i in range(len_수사기관사칭형_text):
    cur_text = voicephishing_수사기관사칭형_text[i]
    cur_text = ast.literal_eval(cur_text)
    voicephishing_수사기관사칭형_text[i] = cur_text
```

```python
# 문자열의 리스트화
for i in range(len_대출사기형_text):
    cur_text = voicephishing_대출사기형_text[i]
    
    # 빈 리스트를 제외한 텍스트가 여전히 리스트 형태일 경우
    if cur_text != []:
        voicephishing_대출사기형_text[i] = cur_text

# 빈 리스트 제거
voicephishing_대출사기형_text = [text for text in voicephishing_대출사기형_text if text != []]

for i in range(len_수사기관사칭형_text):
    cur_text = voicephishing_수사기관사칭형_text[i]
    
    # 빈 리스트를 제외한 텍스트가 여전히 리스트 형태일 경우
    if cur_text != []:
        voicephishing_수사기관사칭형_text[i] = cur_text

# 빈 리스트 제거
voicephishing_수사기관사칭형_text = [text for text in voicephishing_수사기관사칭형_text if text != []]
```

```python
import re

# 피해자/사기범 명칭 제거
for i in range(len(voicephishing_대출사기형_text)):
    cur_text = voicephishing_대출사기형_text[i]
    
    for j in range(len(cur_text)):
        # 피해자 명칭 제거
        if ('피해자:' in cur_text[j]) or ('피해자 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'피해자:', '', cur_text[j])
            cur_text[j] = re.sub(r'피해자 : ', '', cur_text[j])
            
        # 사기범 명칭 제거
        if ('사기범:' in cur_text[j]) or ('사기범 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'사기범:', '', cur_text[j])
            cur_text[j] = re.sub(r'사기범 : ', '', cur_text[j])
            
    # 수정된 텍스트를 다시 리스트에 할당
    voicephishing_대출사기형_text[i] = cur_text
                
for i in range(len(voicephishing_수사기관사칭형_text)):
    cur_text = voicephishing_수사기관사칭형_text[i]
    
    for j in range(len(cur_text)):
        # 피해자 명칭 제거
        if ('피해자:' in cur_text[j]) or ('피해자 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'피해자:', '', cur_text[j])
            cur_text[j] = re.sub(r'피해자 : ', '', cur_text[j])
            
        # 사기범 명칭 제거
        if ('사기범:' in cur_text[j]) or ('사기범 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'사기범:', '', cur_text[j])
            cur_text[j] = re.sub(r'사기범 : ', '', cur_text[j])
            
    # 수정된 텍스트를 다시 리스트에 할당
    voicephishing_수사기관사칭형_text[i] = cur_text
```

```python
from nltk.tokenize import sent_tokenize

# 문장 분리
for i in range(len(voicephishing_대출사기형_text)):
    cur_text = voicephishing_대출사기형_text[i]
    splited_text = []
    
    for j in range(len(cur_text)):
        # 문장 분리 후 길이 저장
        sentences = sent_tokenize(cur_text[j])
        len_splited = len(sentences)
        
        # 길이가 1이면 바로 추가함
        if len_splited == 1:
            splited_text.append(cur_text[j])
        # 길이가 1이 아니면 나눠서 추가함
        else:
            for k in range(len_splited):
                splited_text.append(sentences[k])
            
    # 분리한 문장으로 대체
    voicephishing_대출사기형_text[i] = splited_text
    
    
for i in range(len(voicephishing_수사기관사칭형_text)):
    cur_text = voicephishing_수사기관사칭형_text[i]
    splited_text = []
    
    for j in range(len(cur_text)):
        # 문장 분리 후 길이 저장
        sentences = sent_tokenize(cur_text[j])
        len_splited = len(sentences)
        
        # 길이가 1이면 바로 추가함
        if len_splited == 1:
            splited_text.append(cur_text[j])
        # 길이가 1이 아니면 나눠서 추가함
        else:
            for k in range(len_splited):
                splited_text.append(sentences[k])
            
    # 분리한 문장으로 대체
    voicephishing_수사기관사칭형_text[i] = splited_text
```
