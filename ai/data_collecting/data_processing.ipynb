```python
import os
import pandas as pd

# 파일 경로 설정
file_path = 'crawling.csv'

# 파일이 존재하는지 확인
if os.path.exists(file_path):
    # CSV 파일 불러오기
    df = pd.read_csv(file_path)
    df = df.rename(columns={'내용': 'transcript'})

    # 'label' 열 추가 및 값 할당
    df['label'] = 1

    # 'Unnamed' 컬럼 제거
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

    # 데이터프레임 확인
    print(df.head())

    # 전처리한 데이터프레임을 CSV 파일로 저장
    df.to_csv('/content/preprocessed_data.csv', index=False)
else:
    print(f'Error: 파일 {file_path}이(가) 존재하지 않습니다.')
```

```python
import pandas as pd

# 첫 번째 데이터프레임 읽어오기
yes_vp = pd.read_csv('/content/yesvp.csv')

# 두 번째 데이터프레임 읽어오기
crawling = pd.read_csv('/content/crawling.csv')

# 두 데이터프레임을 단순히 이어붙이기
combined_df = pd.concat([yes_vp, crawling], ignore_index=True)

# 결과 확인
print(combined_df.head())

# 이어붙인 데이터프레임을 CSV 파일로 저장
combined_df.to_csv('/content/combined_data.csv', index=False)

combined_df
```

```python
import pandas as pd
import random

# CSV 파일 경로
csv_path = '/content/df_data_unharmful.csv'

# CSV 파일에서 데이터 읽어오기
df = pd.read_csv(csv_path)

# 317 데이터를 랜덤으로 선택
selected_data = df.sample(n=317, random_state=42)  # random_state는 재현성을 위한 시드 값입니다.

# 선택된 데이터를 새로운 CSV 파일로 저장 (선택은 필요에 따라 다르게 할 수 있습니다.)
selected_data.to_csv('/content/selected_data.csv', index=False)
```

```python
#yes_vp+ no_vp concat
import pandas as pd

# 첫 번째 데이터프레임 읽어오기
yes_vp = pd.read_csv('/content/yesvp_300.csv')

# 두 번째 데이터프레임 읽어오기
novp = pd.read_csv('/content/novp_300.csv')

# 두 데이터프레임을 단순히 이어붙이기
combined_df = pd.concat([yes_vp, novp], ignore_index=True)
# 'Unnamed' 컬럼과 'confidence' 컬럼 제거
combined_df = combined_df.drop(columns=['Unnamed: 0', 'confidence'])

# 결과 확인
print(combined_df.head())

# 이어붙인 데이터프레임을 CSV 파일로 저장
combined_df.to_csv('/content/All_dataset.csv', index=False)

```
